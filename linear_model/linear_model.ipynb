```python
import numpy as np
import matplotlib.pyplot as plt

w = 1.0
w_list = []
mse_list = []

def forward(x):
    return x*w

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2

for w in np.arange(0.0, 4.1, 0.1):
    print(f"w= {w:.1f}")
    l_sum = 0.0
    for x_val, y_val in zip(np.arange(1, 4, 1), np.arange(2, 8, 2)):
        y_prev = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l
        print(f"\tx={x_val:.1f}, y={y_val:.1f}, y_pred={y_prev:.1f}, loss={l:.2f}")
    w_list.append(w)
    mse_list.append(l_sum / 3)
    print(f"MSE= {l_sum / 3:.2f}\n")
# Plotting the results
plt.plot(w_list, mse_list, marker='o')
plt.title('MSE vs Weight (w)')
plt.xlabel('Weight (w)')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid()
plt.xticks(np.arange(0, 4.1, 0.5))
plt.yticks(np.arange(0, 10.1, 1))
plt.xlim(0, 4)
plt.ylim(0, 10)
plt.show()
```

```python
#exercise
import csv
import numpy as np
import matplotlib.pyplot as plt

# -------- CONFIG --------
w = 0.0
years = []
life_expectancy = []
best_mse = float('inf')
best_w = None
w_list = []
mse_list = []

# -------- HELPER FUNCTIONS --------
def forward(x):
    return (x - 2000.0) * w + 76.0  # bias for life expectancy

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2

# -------- READ DATA --------
with open('./Life Expectancy Data.csv', 'r', encoding='UTF-8') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader)  # Skip header
    for row in reader:
        years.append(int(row[1]))
        life_expectancy.append(float(row[3]))

# -------- CALCULATE MSE--------
for w in np.arange(0.0, 1.001, 0.001):
    l_sum = 0.0
    for year, life in zip(years, life_expectancy):
        y_prev = forward(year)
        l = loss(year, life)
        l_sum += l
    w_list.append(w)
    mse = l_sum / len(years)
    mse_list.append(mse)
    if mse < best_mse:
        best_mse = mse
        best_w = w
    print(f"w= {w:.3f}, MSE= {mse:.2f}")

# -------- RESULTS --------
print(f"\nBest weight (w): {best_w:.3f}, Best MSE: {best_mse:.2f}")

# -------- PLOT --------
plt.subplot(1, 2, 1)
plt.plot(years, life_expectancy, 'o', label='Data')
plt.plot(years, [((y - 2000.0) * best_w + 76.0) for y in years], label='Best Model', color='orange')
plt.title('Life Expectancy vs Year')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend()
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(w_list, mse_list)
plt.title('MSE vs Weight (w)')
plt.xlabel('Weight (w)')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid()
plt.xticks(np.arange(0, 1.1, 0.25))
plt.yticks(np.arange(0, 20.1, 2))
plt.xlim(0, 1)
plt.ylim(0, 20)

plt.tight_layout()
plt.show()
    
```

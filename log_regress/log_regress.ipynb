```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable

x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0], [4.0]]))
y_data = Variable(torch.Tensor([[0.], [0.], [1.], [1.]]))

class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1, 1)

    def forward(self, x):
        return F.sigmoid(self.linear(x))
    
model = Model()

criterion = torch.nn.BCELoss(size_average=True)
opt = torch.optim.SGD(model.parameters(), lr = 0.01)

for epoch in range(1, 10001):
    y_pred = model(x_data)

    loss = criterion(y_pred, y_data)
    print(f"epoch: {epoch}\tloss: {loss}")

    opt.zero_grad()
    loss.backward()
    opt.step()

hour_var = Variable(torch.Tensor([[1.]]))
print(f"Predict 1H: {model(hour_var).data[0][0] > 0.5}")
hour_var = Variable(torch.Tensor([[7.]]))
print(f"Predict 7H: {model(hour_var).data[0][0] > 0.5}")
```

```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable

x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0], [4.0]]))
y_data = Variable(torch.Tensor([[0.], [0.], [1.], [1.]]))

class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(1, 1)

    def forward(self, x):
        return F.hardtanh(self.linear(x), min_val=0, max_val=1)
    
model = Model()

criterion = torch.nn.BCELoss(size_average=False)
opt = torch.optim.SGD(model.parameters(), lr = 0.001)

for epoch in range(1, 10001):
    y_pred = model(x_data)

    loss = criterion(y_pred, y_data)
    print(f"epoch: {epoch}\tloss: {loss}")

    opt.zero_grad()
    loss.backward()
    opt.step()

hour_var = Variable(torch.Tensor([[1.]]))
print(f"Predict 1H: {model(hour_var).data[0][0] > 0.5}")
hour_var = Variable(torch.Tensor([[7.]]))
print(f"Predict 7H: {model(hour_var).data[0][0] > 0.5}")
```

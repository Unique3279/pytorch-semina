```python
import torch
import time
from torch.autograd import Variable

x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0]]))
y_data = Variable(torch.Tensor([[2.0], [4.0], [6.0]]))

class LinearModel(torch.nn.Module):
    def __init__(self):
        super(LinearModel, self).__init__()
        self.linear = torch.nn.Linear(1, 1) # 1 input and 1 output

    def forward(self, x):
        return self.linear(x)
    
model = LinearModel()

# set loss function and optimizer
criterion = torch.nn.MSELoss(size_average=False)

# SGD : 일부 데이터의 미니배치로 가중치를 업데이트하는 방법
# BGD보다 빨리 수렴하지만 다소 불안정함, 업데이트를 반복하면 결국 수렴
opt = torch.optim.SGD(model.parameters(), lr=0.01) # lr = learning rate

train_start = time.time()
for epoch in range(1, 10001):
    y_pred = model(x_data)  # Forward pass
    loss = criterion(y_pred, y_data)    # Compute loss
    print(f"epoch: {epoch}, loss: {loss.item()}")

    opt.zero_grad()  # Zero the gradients
    loss.backward()  # Backward pass
    opt.step()      # Update weights 
    if loss < 0.001:
        print("Early stopping at epoch:", epoch)
        break
train_end = time.time()
print(f"Training time: {(train_end - train_start):.2f} s")

# Test the model
test_data = Variable(torch.Tensor([[4.0]]))
predicted = model(test_data)
print("Prediction (after training)", predicted.data[0][0])
```

```python
# practice 5-1
import torch
from torch.autograd import Variable
import time

x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0]]))
y_data = Variable(torch.Tensor([[2.0], [4.0], [6.0]]))

class LinearModel(torch.nn.Module):
    def __init__(self):
        super(LinearModel, self).__init__()
        self.linear = torch.nn.Linear(1, 1) # 1 input and 1 output

    def forward(self, x):
        return self.linear(x)
    
model = LinearModel()

# set loss function and optimizer
criterion = torch.nn.MSELoss(size_average=False)
opt = torch.optim.Adam(model.parameters(), lr=0.01) # lr = learning rate

train_start = time.time()
for epoch in range(1, 10001):
    y_pred = model(x_data)  # Forward pass
    loss = criterion(y_pred, y_data)    # Compute loss
    print(f"epoch: {epoch}, loss: {loss.item()}")

    opt.zero_grad()  # Zero the gradients
    loss.backward()  # Backward pass
    opt.step()      # Update weights 
    if loss < 0.001:
        print("Early stopping at epoch:", epoch)
        break
train_end = time.time()
print(f"Training time: {(train_end - train_start):.2f} s")

# Test the model
test_data = Variable(torch.Tensor([[4.0]]))
predicted = model(test_data)
print("Prediction (after training)", predicted.data[0][0])
```
